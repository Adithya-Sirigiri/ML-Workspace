## Stem plot

```python
from matplotlib import pyplot as plt
x=[1,2,3,4,5,6]
y=[2,5,4,3,6,8]

plt.stem(x,y)
plt.show()
```

```python
from matplotlib import pyplot as plt
x=[1,2,3,4,5,6]
y=[2,5,4,3,6,8]

plt.xlabel("Random")
plt.ylabel("Frequency values")
plt.title("Stem plot")
plt.stem(x,y)
plt.show()
```

```python
from matplotlib import pyplot as plt
x=[1,2,3,4,5,6]
y=[2,5,4,3,6,8]

plt.xlabel("Random")
plt.ylabel("Frequency values")
plt.title("Stem plot")
plt.stem(x,y,linefmt=':')
plt.show()
```

```python
from matplotlib import pyplot as plt
x=[1,2,3,4,5,6]
y=[2,5,4,3,6,8]

plt.xlabel("Random")
plt.ylabel("Frequency values")
plt.title("Stem plot")
plt.stem(x,y,linefmt=':',markerfmt='g+')
plt.show()
```

```python
from matplotlib import pyplot as plt
x=[1,2,3,4,5,6]
y=[2,5,4,3,6,8]

plt.xlabel("Random")
plt.ylabel("Frequency values")
plt.title("Stem plot")
plt.stem(x,y,linefmt=':',markerfmt='g+',basefmt='m')
plt.show()
```

```python
from matplotlib import pyplot as plt
x=[1,2,3,4,5,6]
y=[2,5,4,3,6,8]

plt.xlabel("Random")
plt.ylabel("Frequency values")
plt.title("Stem plot")
plt.stem(x,y,linefmt=':',markerfmt='g+',basefmt='m',orientation='horizontal')
plt.show()
```

```python
from matplotlib import pyplot as plt
x=[1,2,3,4,5,6]
y=[2,5,4,3,6,8]

plt.xlabel("Random")
plt.ylabel("Frequency values")
plt.title("Stem plot")
plt.stem(x,y,linefmt=':',markerfmt='g+',basefmt='m',use_line_collection=False)
plt.show()
```
