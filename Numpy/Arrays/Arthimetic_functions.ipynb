```python
import numpy as np
var=np.array([1,3,5,7])
print(np.max(var))
print(np.min(var))
```

```python
var_1=np.array([1,3,5,7])
print(np.max(var_1),np.argmax(var_1))
print(np.min(var_1),np.argmin(var_1))
```

```python
var_2=np.array([1,3,5,7])
print(np.sin(var_2))
print(np.cos(var_2))
print(np.cumsum(var_2))
```

```python
var_2=np.array([1,3,5,7,16])
print(np.sqrt(var_2))
```

```python
var_2=np.array([[1,3,5,7],[2,4,6,8]])
print(np.max(var_2,axis=0))
print(np.max(var_2,axis=1))
print(np.min(var_2,axis=0))
print(np.min(var_2,axis=1))
```
