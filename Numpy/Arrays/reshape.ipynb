```python
import numpy as np
var=np.array([2,4,6,7,9])
print(var)
print(var.ndim)
print(var.shape)
```

```python
var1=np.array([[2,4,6,7,9],[1,3,6,4,0]])
print(var1)
print(var1.ndim)
print(var1.shape)
print(var1.reshape(5,2))
```

```python
var3=np.array([[[2,4,6,7,9],[1,3,6,4,0],[4,6,2,9,7]]])
print(var3)
print(var3.ndim)
print(var3.shape)
print(var3.reshape(3,5,1))
print()
print(var3.reshape(1,5,3))
```

1-D to 2-D

```python
var=np.array([2,4,6,7,9,0])
print(var)
print(var.ndim)
print(var.shape)
print()
print()
print(var.reshape(3,2))
```

1-D to 3-D

```python
var=np.array([2,4,6,7,9,0])
print(var)
print(var.ndim)
print(var.shape)
print()
print()
print(var.reshape(1,3,2))
print()
print(var.reshape(2,3,1))
```

Convert anything into 1-D

```python
var=np.array([2,4,6,7,9,0])
print(var)
print(var.ndim)
print(var.shape)
print()
print()
print(var.reshape(1,3,2))
print()
print(var.reshape(-1))
```
