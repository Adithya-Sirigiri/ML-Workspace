```python
import numpy as np
var=np.array([1,3,5,7,9])
var1=np.array([2,4,6,8])
print(var+var1)
```

```python
var2=np.array([[1,3,5,7,9],[2,4,6,8,0]])
var3=np.array([[2],[1]])
print(var2+var3)
```
