#TYPES OF ARRAY

##1-D array

```python
import numpy as np
arr_1=np.array([2,3,5,6])
print(arr_1)
print(arr_1.ndim)
print(arr_1.shape)
```

##2-D array

```python
arr_2=np.array([[2,3,5,6],[1,9,8,7]])
print(arr_2)
print(arr_2.ndim)
print(arr_2.shape)
```

##3-D array

```python
arr_3=np.array([[[2,3,5,6],[1,7,8,9],[3,6,4,8]]])
print(arr_3)
print(arr_3.ndim)
print(arr_3.shape)
```

##n dimensional array

```python
arn=np.array([1,2,3,4],ndim=10)
print(arn)
```
