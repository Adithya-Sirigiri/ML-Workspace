```python
import numpy as np
var=np.array([1,3,5,7])
var_1=np.array([2,4,6,8])
print(var+var_1)
print(var-var_1)
print(var*var_1)
print(var/var_1)
print(var%var_1)
print(var**var_1)
print(1/var)
print(1/var_1)
```

```python
var=np.array([1,3,5,7])
var_1=np.array([2,4,6,8])
print(np.add(var,var_1))
print(np.subtract(var,var_1))
print(np.multiply(var,var_1))
print(np.divide(var,var_1))
print(np.mod(var,var_1))
print(np.power(var,var_1))
print(np.reciprocal(var))
print(np.reciprocal(var_1))
```

```python
var=np.array([[1,3,5,7],[2,4,6,8]])
var_1=np.array([[2,4,6,8],[1,4,5,7]])
print(np.add(var,var_1))
print(np.subtract(var,var_1))
print(np.multiply(var,var_1))
print(np.divide(var,var_1))
print(np.mod(var,var_1))
print(np.power(var,var_1))
```
