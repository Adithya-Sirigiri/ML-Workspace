##Numpy array using numpy function
##Special numpy arrays

##zeros array

```python
import numpy as np
arr=np.zeros(4)
print(arr)
```

```python
arr=np.zeros((4,3))
print(arr)
```

##ones array

```python
ar_1=np.ones(4)
ar_2=np.ones((4,3))
print(ar_1)
print()
print(ar_2)
```

##empty array

```python
e=np.empty((4))
f=np.empty((4,3))
print(e)
print()
print(f)
```

##for a range

```python
ra=np.arange(0,5)
print(ra)
```

##linscape

```python
li=np.linspace(0,12,5)
print(li)
```

##diagonal elements with one's

```python
dia=np.eye(4)
print(dia)
```
